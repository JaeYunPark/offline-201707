```python
# 정렬이 필요하다면
for i in sorted({1, 3, 4, 5, 1, 4, 3, 1}, reverse=True):
    print(i)
```

```python
# 중복을 제거한 리스트
list_numbers = [1, 1, 2, 1, 1, 2, 2, 3, 1, 2, 3]
result = []
for i in list_numbers:
    if i not in result:
        result.append(i)
print(result)

print(list(set(list_numbers)))
```
